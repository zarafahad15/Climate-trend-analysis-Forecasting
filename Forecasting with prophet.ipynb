```python
import pandas as pd

# load dataset
df = pd.read_csv("temperature.csv")

# basic cleanup
df["dt"] = pd.to_datetime(df["dt"])
df = df[df["AverageTemperature"].notnull()]

# set date as index
df = df.set_index("dt").sort_index()

# resample to yearly mean temperature
yearly = df["AverageTemperature"].resample("Y").mean()
yearly.head()
```
